```python
import tensorflow as tf
import tensorflow.keras as keras
from sklearn.model_selection import train_test_split

y_labels = train.loc[:, 'class']
X_data = train.loc[:, ['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm']]
```

```python
labels = {
    'Iris-setosa': [1,0,0],
    'Iris-versicolor':[0,1,0],
    'Iris-virginica':[0,0,1]
}

y_nums = np.array(list(map(lambda v: labels[v], y_labels)))
X_data = np.array(X_data)
columns = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X_data, y_nums, test_size=0.3)
```

```python
Dense = keras.layers.Dense
model = keras.models.Sequential()
model.add(Dense(10, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])
```

```python
model.fit(X_data, y_nums, batch_size=20, epochs=300)
```

```python
X_test_data = test.loc[:, ['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm']]
X_test_data = np.array(X_test_data)
```

```python
pred = model.predict(X_test_data)
pred = pd.DataFrame(pred, columns=columns)
pred.head()
```
